# Example of environment simulation

```python
import numpy as np
import matplotlib.pyplot as plt
from gxr.envir.functions.utility import UtilLinSqrt, UtilLinRoot
```

## Linear-square root utility

$$
u(x) = \begin{cases}
    bx\qquad&\text{for $x < 0$} \\
    2\sqrt(x + 1/b^2) - 2/b\qquad&\text{for $x >= 0$}
\end{cases}
$$
$$
\frac{du}{dt} = \begin{cases}
    b\qquad&\text{for $x < 0$} \\
    \frac{1}{\sqrt{x + 1/b^2}}\qquad&\text{for $x >= 0$}
\end{cases}
$$

```python
x = np.linspace(-2, 10, 500)
B = [.5, 1, 2, 4, 8]

fig, ax = plt.subplots(figsize=(5, 5))
ax.axvline(0, ls="--", color="gray")
ax.axhline(0, ls="--", color="gray")
for b in B:
    util = UtilLinSqrt(b)
    ax.plot(x, util(x), label=rf"$b={b}$")
ax.legend()
```

## Linear-arbitrary root utility

$$
u(x) = \begin{cases}
    bx\qquad&\text{for $x < 0$} \\
    b\left[
        \left(x + c^{\frac{c}{1-c}}\right)^{\frac{1}{c}}
        - c^{\frac{1}{1-c}}
    \right]
\end{cases}
$$
$$
\frac{du}{dt} = \begin{cases}
    b\qquad&\text{for $x < 0$} \\
    \frac{b}{c}\left(x + c^{\frac{c}{1-c}}\right)^{\frac{1-c}{c}}
\end{cases}
$$

```python
x = np.linspace(-2, 10, 500)
B = [1, 4]
C = [1, 2, 4]

fig, ax = plt.subplots(figsize=(5, 5))
ax.axvline(0, ls="--", color="gray")
ax.axhline(0, ls="--", color="gray")
for b in B:
    for c in C:
        util = UtilLinRoot(c, b)
        ax.plot(x, util(x), label=rf"$c={b}, b={c}$")
ax.legend()
```
